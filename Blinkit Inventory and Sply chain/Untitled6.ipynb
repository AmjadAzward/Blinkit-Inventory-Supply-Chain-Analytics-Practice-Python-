```python
import numpy as np #scientific and mathematical calculation
import pandas as pd #data manipulation
import seaborn as sns #Visualization
import matplotlib.pyplot as plt #Visualization
```

```python
orders = pd.read_csv("Orders.csv") #read file within pandas dataframe
orders
```

```python
orders.info()  #to see details about columns, rows, datatypes and non null values
```

```python
#Fixing datatypes

#if not datettime
#Suppose quantity was a string (text)

#orders["Quantity"] = orders["Quantity"].astype("int")

#if datetime

orders["OrderDateTime"] = pd.to_datetime(orders["OrderDateTime"]) #convert object to datetime format
```

```python
orders["ExpectedDeliveryTime"] = pd.to_datetime(orders["ExpectedDeliveryTime"])
orders["ActualDeliveryTime"] = pd.to_datetime(orders["ActualDeliveryTime"])
```

```python
#to subtract actual delivery time and order time and get the output in minutes.
#dt stands for datetime which is a module in python that can handle dates and time
orders["Actual_Delivery_in_mins"] = (orders["ActualDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
orders["Expected_Delivery_in_mins"] = (orders["ExpectedDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
#to see distinct values in a column
orders["StoreID"].unique()
```

```python
orders["Diff_bw_Expected_Actual"] = orders["Expected_Delivery_in_mins"] - orders['Actual_Delivery_in_mins']
orders
```

```python
#Loc allows us to work with indexes
#
orders.loc[(orders["Diff_bw_Expected_Actual"] == 0), "order_status"] = "On Time"
orders.loc[(orders["Diff_bw_Expected_Actual"] < 0), "order_status"] = "Delayed"
```

```python
plt.figure(figsize= (16,4)) #increase the chart size
sns.countplot(x = "StoreID" , data = orders , hue = "order_status") #hue allows us to provide legend
plt.xticks(rotation = 90) #to rotate x axis labels to 90 degrees
plt.show() #to display the visual
```
